#Vigenere Cypher

This cypher uses a spcific key to both encrypt and decrypt the message

```python
alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
```

```python
def generateKey(string, key):
  key = list(key)
  if len(string) == len(key):
    return(key)
  else:
    for i in range(len(string)-len(key)):
      key.append(key[i])
    return"" . join(key)
```

Here is the funtion for the user's key

1. Creates a function but with the argument sting and key
2. Key = The users key but now in a list ex: T,e,s,t
3. If the length of the string is equal to the legth of the User generated key, it would return the key
4. Else, If the length of string does not equal the length of key, find the index in the range of the length of the string and subtract that from the length of key
5. this appends the letters from the previous line
6. It is then returned to form the key

```python
def vig_encode(string, key):
  cipher_text = []
  for i in range(len(string)):
    m_num = alpha.index(string[i])
    k_num = alpha.index(key[i])
    e_num = (m_num+k_num) % 26
    cipher_text.append(alpha[e_num])
  return "" . join(cipher_text)
```

Here is the encryption of the given word

1. Creates a function with the arguments string and key

```python
def vig_decode(string, key):
  cipher_text = []
  for i in range(len(string)):
    m_num = alpha.index(string[i])
    k_num = alpha.index(key[i])
    e_num = (m_num-k_num) %26
    cipher_text.append(alpha[e_num])
  return "" . join(cipher_text)
```
