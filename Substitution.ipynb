##Substitution Cypher

This cypher goes one step over the Ceaser cypher and scrambles the alphabet as well as shifting the letters

```python
alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
```

```python
def sub_encode(text, altalpha):
  answer_list=[]
  for letter in text:
    ind = alpha.index(letter)#Find the index of the alphabet
    new_letter = altalpha[ind] #Find the index of the alt alpha that matches the index of alpha
    answer_list.append(new_letter)
  return "".join(answer_list)
```

This is the coding for the *encode*

- Create a list (called answer_list) thats empty
- for letter from the input in text, create a variable(ind)
- ind = the alpha(alphabet) index of the letter
- (new variable) new_letter = the index of the alt alpha that matches the index of alpha
- append the new_letter to the list we created(answer_list)

```python
def sub_decode(text, altalpha):
  answer_list=[]
  for letter in text:
    ind = altalpha.index(letter)
    new_letter=alpha[ind]
    answer_list.append(new_letter)
  return "".join(answer_list)
```

This is the coding for the *decode*

- Create a list thats empty
- for letter from the encode in text, create a variable(a different ind) 
- Ind = the alt alpha (re-lettered alphabet) index of the letter
- (new variable) new_letter = the index of the alpha(alphabet) that matches the index of altalpha
append the letter(new_letter) into the answer_list

```python
test = "WORLDSHELLO"
cipher_alphabet = "WJKUXVBMIYDTPLHZGONCRSAEFQ"
enc = sub_encode(test, cipher_alphabet)
dec = sub_decode(enc, cipher_alphabet)
print(enc)
print(dec)
```
